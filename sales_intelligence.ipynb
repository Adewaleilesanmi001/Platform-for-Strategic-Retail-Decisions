#### import Libaries


```python
import pandas as pd
import numpy as np
```

#### Extration of data


```python
df = pd.read_csv(r'./Sales Dataset.csv')
```

```python
df.info()
```

#### Data cleaning 

```python
### Formating data column names

df.columns = (
    df.columns.str.strip()
               .str.lower()
               .str.replace('-', '_', regex=True)
               .str.replace(' ', '_', regex=True)
)
```

```python
df.columns
```

```python
### Checking the information of the Data 

df.info()
```

```python
df = df.drop_duplicates().reset_index(drop=True) ### Droping duplicate if any 
df.info()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='%d/%m/%Y', errors='coerce' ) #formarting the order_date column to a date and time dataType
df['year_month'] = pd.to_datetime(df['year_month'], format='%Y-%m', errors='coerce')  #formarting the year_month column to a date and time dataType
df.info()
```

```python
df['match'] = (df['order_date'].dt.year == df['year_month'].dt.year) & \
              (df['order_date'].dt.month == df['year_month'].dt.month) #checking if the order_date == year_month

print(df['match'].drop_duplicates()) # droroping duplicate in the compare to see if we have a false but all are True so the both data are same so i can drop the year month column.
```

```python
df.info()
```

```python
df = df.drop('match', axis='columns') # droping the match testing column 
df = df.drop('year_month', axis='columns') # droping the duplicated column
```

```python
df.info()
```

```python
### Spliting the name coulumn into firt name and last name base on 1NF

df[['first_name', 'last_name']] = df['customername'].str.split(' ', n=1, expand=True)
```

#### Creating a Product table 

```python
df_products = df[['category', 'sub_category']].copy().drop_duplicates().reset_index(drop=True)
df_products['product_id'] = df_products.index + 1
df_products = df_products[['product_id', 'category', 'sub_category']]
df_products
```

#### Creating a customer table 

```python
df_customers = (
    df[['first_name', 'last_name', 'state', 'city']]
    .copy()
    .drop_duplicates()
    .reset_index(drop=True)
)
df_customers['customer_id'] = df_customers.index + 1
# df_customers[['first_name', 'last_name']] = df_customers['customername'].str.split(' ', n=1, expand=True)
df_customers = df_customers[['customer_id', 'first_name', 'last_name', 'state', 'city']]
df_customers


```

#### Creating an order table 

```python
df_orderss = (
    df[['order_id', 'order_date', 'amount', 'paymentmode', 'quantity', 'first_name', 'last_name', 'sub_category', 'state', 'city', 'profit']]
    .merge(df_customers[['customer_id', 'first_name', 'last_name', 'state', 'city']], 
           left_on=['first_name', 'last_name', 'state', 'city'], 
           right_on=['first_name', 'last_name', 'state', 'city'], 
           how='left', validate='many_to_one')
    .merge(df_products[['product_id', 'sub_category']], on='sub_category', how='left', validate='many_to_one')
).copy().drop_duplicates().reset_index(drop=True)
df_orderss['cost_per_unit'] = np.where(
    df_orderss['quantity'] != 0,
    ((df_orderss['amount'] - df_orderss['profit']) / df_orderss['quantity']).round(2),
    np.nan
)
df_orderss['price'] = (df_orderss['amount'] / df_orderss['quantity']).round(2)
df_orderss.insert(0, 'order_pk', df_orderss.index + 1)

df_orders = df_orderss[['order_pk', 'order_id', 'order_date', 'amount', 'paymentmode', 'quantity', 'customer_id', 'product_id', 'price']]

df_orders

```

#### Creating a revenue table 

```python

df_revenue = (
    df_customers[['customer_id']]
            .merge(df_orderss[['order_pk', 'order_id',
       'customer_id', 'product_id', 'profit', 'cost_per_unit']], on='customer_id', how='left')
            .merge(df_products[[ 'product_id']], on='product_id', how='left')
                  ).copy().drop_duplicates()

df_revenue['revenue_id'] = df_revenue.index + 1
df_revenue = df_revenue[['revenue_id', 'customer_id', 'order_pk', 'product_id', 'cost_per_unit', 'profit']]

df_revenue

```

#### Creating a Fast table 

```python


df_fact_table = (
    df_customers[['customer_id']]
            .merge(df_orderss[['order_pk', 'order_id', 'order_date', 'amount', 'paymentmode', 'quantity',
       'customer_id', 'product_id', 'price', 'profit', 'cost_per_unit']], on='customer_id', how='left')
            .merge(df_products[['category', 'sub_category' , 'product_id']], on='product_id', how='left')
          #   .merge(df[['profit', 'amount', 'order_date','order_id', 'first_name', 'last_name', 'state', 'city']], on=['first_name', 'last_name', 'state', 'city'],  how='left', validate='many_to_many'). reset_index(drop=True) 
                 ).copy().drop_duplicates().reset_index(drop=True)
df_fact_table['month'] = df_fact_table['order_date'].dt.strftime('%B')
df_fact_table['year'] = df_fact_table['order_date'].dt.year
df_fact_table['transaction_id'] = df_fact_table.index + 1
# df_fact_table = df_fact_table[['order_id', 'customer_id', 'city', 'category', 'sub_category', 'paymentmode', 'order_date', 'amount', 'profit', 'quantity']]
df_fact_table = df_fact_table[['transaction_id', 'customer_id', 'order_pk', 'order_date', 'amount', 'paymentmode',
       'quantity', 'product_id', 'price', 'profit', 'cost_per_unit',
        'month', 'year']]
df_fact_table


```

#### Importing Clean table into a CSV file

```python
## Writing the clean data into a CSV file 

df_fact_table.to_csv('df_fact_table.csv', index=False)
df_revenue.to_csv('df_revenue.csv', index=False)
df_orders.to_csv('df_orders.csv', index=False)
df.to_csv('clean_sales_data.csv', index=False)
df_products.to_csv('df_products.csv', index=False)
df_customers.to_csv('df_customers.csv', index=False)
```

#### Analysing profit and loss and profit margin by customer

```python
profit_loss = (
    df_customers[['customer_id', 'city']]
            .merge(df_fact_table[['customer_id', 'amount', 'profit', 'product_id']], on='customer_id', how='left')
            .merge(df_products[['product_id', 'category']], on='product_id', how='left')
            
    .groupby(['customer_id', 'city', 'category'])
    .agg(total_sales = ('amount', 'sum'),
         total_profit = ('profit', 'sum'))
         .reset_index()
)
profit_loss['status'] = ''
profit_loss['profit_margin_percentage'] = ''

for i in range(len(profit_loss['total_profit'])):
    if profit_loss.loc[i, 'total_profit'] > 0:
        profit_loss.loc[i, 'status'] = 'Profit'
    else:
        profit_loss.loc[i, 'status'] = 'Loss'

for i in range(len(profit_loss['total_profit'])):
    profit_loss.loc[i, 'profit_margin_percentage'] = ((profit_loss.loc[i, 'total_profit']/profit_loss.loc[i, 'total_sales']) *100).round(2)
        

profit_loss


```

#### Analysing month;y profit by category

```python
month_profit_category = (
    df_fact_table[['order_date', 'product_id', 'profit']].merge(df_products[['product_id', 'category']], on='product_id', how='left')
    .groupby(['order_date', 'category'])
    ['profit'].sum()
            .reset_index()
            .sort_values('profit', ascending=False)
                ).reset_index(drop=True)
month_profit_category
```

#### Analysing Profit made by month and year

```python
monthly_profit = (
    df_fact_table[['month', 'year', 'profit']]
    .groupby(['month', 'year'])['profit']
      .sum()
      .reset_index()
      .sort_values(['month', 'year'])
      .reset_index(drop=True)
)

monthly_profit
```

#### Analysing Profit by customers

```python
top_customers = (
    df_customers[['customer_id', 'first_name', 'last_name']]
    .merge(df_fact_table[['customer_id', 'profit']], on='customer_id').reset_index(drop=True)
    .groupby(['first_name', 'last_name'])['profit']
              .sum()
              .reset_index()
              .sort_values('profit', ascending=False)
              .reset_index(drop=True)
)
top_customers
```

#### Analysing quanty by category

```python
quantity_per_category  = (
    df_products[['category','product_id']].merge(df_fact_table[['product_id', 'quantity']], )
    .groupby([ 'category'])['quantity']
              .sum()
              .reset_index()
              .sort_values('quantity', ascending=False)
              .reset_index(drop=True)
)
quantity_per_category
```

#### Analysing Profit by city from the highest

```python
city_with_low_profit = (
    df_customers[['customer_id', 'city']].merge(df_fact_table[['customer_id', 'profit']])
    .groupby([ 'city'])['profit']
              .sum()
              .reset_index()
              .sort_values('profit', ascending=True)
              .reset_index(drop=True)
)

city_with_low_profit
```

#### Analysing city with the highest revenue over 95%

```python
Cities_with_revenue= df_customers[['customer_id', 'city']].merge(df_fact_table[['customer_id', 'cost_per_unit', 'quantity']], on='customer_id', how='left')
# grouped_revenue = df.groupby('Product')['Revenue'].sum()
Cities_with_revenue['revenue'] = Cities_with_revenue['cost_per_unit']*Cities_with_revenue['quantity']


# Compute total revenue per city
city_revenue = Cities_with_revenue.groupby('city', as_index=False)['revenue'].sum()

# Calculate the 95th percentile of revenue
threshold = city_revenue['revenue'].quantile(0.95)

# Filter cities with revenue above the 95th percentile
top_cities = city_revenue[city_revenue['revenue'] > threshold]

# Display the result
top_cities
```

#### Analysing revenue made by month and year

```python


monthly_sales = df_fact_table[['product_id', 'month', 'year']].reset_index(drop=True)
monthly_sales['revenue'] = Cities_with_revenue[['revenue']]

# # Aggregate total revenue per product per month
monthly_sales = monthly_sales.groupby(['product_id', 'month', 'year'])\
                    ['revenue'].sum().reset_index()\
                    .sort_values(['month', 'year'])

monthly_sales
```

#### Analysing average monthly revenue per product

```python
 # Calculate average monthly revenue per product
avg_monthly_sales = monthly_sales.groupby('product_id')['revenue'].mean().round(2).reset_index()
avg_monthly_sales.rename(columns={'revenue': 'avg_monthly_revenue'}, inplace=True)

avg_monthly_sales
```

```python
from sqlalchemy import create_engine, Column, Integer, String, Float, ForeignKey, Date, text
from sqlalchemy.orm import declarative_base, sessionmaker, relationship
Base = declarative_base()

```

#### Connecting the DB

```python
engine = create_engine("postgresql+psycopg2://postgres:@localhost:5432/sales_intelligence")
print("✅ SQLAlchemy engine created successfully!")
```

#### Creating of Schema

```python
schema_name = "sales_info"

with engine.connect() as conn:
    conn.execute(text(f"CREATE SCHEMA IF NOT EXISTS {schema_name}"))
    conn.commit()
```

#### Drop tables if they exist

```python
with engine.connect() as conn:
    conn.execute(text("DROP TABLE IF EXISTS sales_info.products CASCADE"))
    conn.execute(text("DROP TABLE IF EXISTS sales_info.customers CASCADE"))
    conn.execute(text("DROP TABLE IF EXISTS sales_info.orders CASCADE"))
    conn.execute(text("DROP TABLE IF EXISTS sales_info.revenue CASCADE"))
    conn.commit()
```

#### Creating tables

```python
class Product(Base):
    __tablename__ = 'products'
    __table_args__ = {"schema": 'sales_info'}
    product_id = Column(Integer, primary_key=True, autoincrement=True)
    category = Column(String(50))
    sub_category = Column(String(50))
    
    # # Relationship to orders
    # orders = relationship("Order", back_populates="product")


# Base.metadata.drop_all(engine, checkfirst=True)  # checkfirst prevents errors if no tables exist

Base.metadata.create_all(engine, checkfirst=True)
print("✅ Tables created successfully!")
```

```python
class Customer(Base):
    __tablename__ = 'customers'
    __table_args__ = {"schema": 'sales_info'}
    customer_id = Column(Integer, primary_key=True, autoincrement=True)
    first_name = Column(String(50))
    last_name = Column(String(50))
    city = Column(String(50))
    state = Column(String(50))
    
    # Relationship to orders
    # orders = relationship("Order", back_populates="customer")

Base.metadata.create_all(engine, checkfirst=True)
print("✅ Tables created successfully!")
```

```python
class Orders(Base):
    __tablename__ = 'orders'
    __table_args__ = {"schema": 'sales_info'}
    order_pk = Column(Integer,  primary_key=True,  autoincrement=True)
    order_id = Column(String(8))
    order_date = Column(Date)
    amount = Column(Integer)
    paymentmode = Column(String(50))
    quantity = Column(Integer)
    customer_id = Column(Integer, ForeignKey('sales_info.customers.customer_id'))
    product_id = Column(Integer, ForeignKey('sales_info.products.product_id'))
    price = Column(Float)

Base.metadata.create_all(engine, checkfirst=True)
print("✅ Tables created successfully!")
```

```python
class Revenue(Base):
    __tablename__ = 'revenue'
    __table_args__ = {"schema": 'sales_info'}
    revenue_id = Column(Integer, primary_key=True, autoincrement=True)
    customer_id = Column(Integer, ForeignKey('sales_info.customers.customer_id'))
    order_pk = Column(Integer, ForeignKey ('sales_info.orders.order_pk'))
    product_id = Column(Integer, ForeignKey('sales_info.products.product_id'))
    cost_per_unit = Column(Float)
    profit = Column(Integer)
    
    

Base.metadata.create_all(engine, checkfirst=True)
print("✅ Tables created successfully!")
```
